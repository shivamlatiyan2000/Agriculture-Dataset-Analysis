```python
#now seperating my features and target
X=df.drop('label',axis=1)
X
```

```python
Y=df['label']
Y
```

```python
#applying feature scaling on independent data--> X
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
X_new=mn.fit_transform(X)
```

```python
X_new
```

```python
#in order to confirm
plt.scatter(df['N'],df['P'])
plt.show()
```

```python
#split the data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_new,Y,test_size=0.20)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
#KNN model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
x=[]
scores=[]
for k in range(1,35,2):
    knn=KNeighborsClassifier(n_neighbors=k,metric='minkowski',weights='uniform',algorithm='auto')
    scores.append(cross_val_score(knn,X_train,Y_train,cv=5).mean())
    x.append(k)
```

```python
plt.plot(x,scores)
plt.grid()
plt.show()
```

```python
xt=[]
st=[]
for k in range(1,35,2):
    knn=KNeighborsClassifier(n_neighbors=k,metric='minkowski',weights='uniform',algorithm='auto')
    st.append(cross_val_score(knn,X_test,Y_test,cv=5).mean())
    xt.append(k)
```

```python
plt.plot(xt,st)
plt.grid()
plt.show()
```

```python
#final model
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1,weights='uniform',metric='minkowski',algorithm='auto')
knn.fit(X_train,Y_train)
```

```python
Y_predict=knn.predict(X_test)
Y_predict
```

```python
print(cross_val_score(knn,X_test,Y_test,cv=5).mean())
```

```python
print(cross_val_score(knn,X_train,Y_train,cv=5).mean())
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_predict))
```
